```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = f"https://www.amazon.in/s?k=bags&page=---&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283"

products_list = []

api_key = input("Enter your API key of ScraperAPI: ")
```

```python
def get_page(url):
    # print("scraping page", url)
    payload = {'api_key': api_key, 'url': url}
    r = requests.get('https://api.scraperapi.com/', params=payload)
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup
```

```python
# find all products details in a page
def get_products(soup):

    products = soup.find_all('div', {'data-component-type': 's-search-result'})

    # find product details
    for product in products:
        try:
            temp_dict = {}
            # url
            temp_dict["URL"] = str("https://amazon.in" + product.find('a', {'class': 'a-link-normal s-no-outline'})['href'])
            # name
            temp_dict["Name"] = product.find('span', {'class': 'a-size-medium a-color-base a-text-normal'}).text
            # price
            temp_dict["Price"] = product.find('span', {'class': 'a-price-whole'}).text
            # rating
            temp_dict["Rating"] = float(product.find('span', {'class': 'a-icon-alt'}).text.split(' ')[0])
            # number of reviews
            temp_dict["Reviews"] = (product.find('span', {'class': 'a-size-base s-underline-text'}).text)

            # append product to products list
            products_list.append(temp_dict)
            
            # delete temp_dict
            del temp_dict
        except:
            product_url = str("https://amazon.in" + product.find('a', {'class': 'a-link-normal s-no-outline'})['href'])
            print("error while parsing product", product_url)
            pass
        
    
```

```python
# get all pages
for i in range(1, 21):
    soup = get_page(URL.replace('---', str(i)))
    get_products(soup)
    print("Page " + str(i) + " scraped successfully")
```

```python
def get_product_details(product_url: str, product_details: dict):
    soup = get_page(test_url)
    # print(soup)

    try:
        try:
            # product description
            product_details["Description"] = soup.find('div', {'id': 'feature-bullets'}).text
        except:
            print("error while parsing product description", product_url)
            product_details["Description"] = "Na"

        try:
            # product ASIN
            product_details["ASIN"] = soup.find('div', {'id': 'productDetails_db_sections'}).find('td', {"class": "a-size-base prodDetAttrValue"}).text.strip()
        except:
            print("error while parsing product ASIN", product_url)
            product_details["ASIN"] = "Na"

        try:
            # product manufacturer
            # print("----------------------", soup.find('table', {'id': 'productDetails_techSpec_section_1'}).find_all('tr')[1].find("td").text.strip())
            product_details["Manufacturer"] = soup.find('table', {'id': 'productDetails_techSpec_section_1'}).find_all('tr')[1].find("td").text.strip()
        except:
            print("error while parsing product manufacturer", product_url)
            product_details["Manufacturer"] = "Na"

        try:
            # more product description
            product_details["More Description"] = "".join(soup.find('div', {'class': 'aplus-v2 desktop celwidget'}).text.strip().split('\n'))
        except:
            print("error while parsing product more description", product_url)
            product_details["More Description"] = "Na"
            
    except Exception as e:
        print("error: ", e)
        print("error while parsing product details", product_url)
        pass

    return product_details
    
```

```python
final_products_list = []
```

```python
# get product details
for (i, product) in enumerate(products_list[:220]):
    product_details = get_product_details(product['URL'], product)
    final_products_list.append(product_details)
    print(f"{i+1}. product details fetched successfully", product['URL'])
    # print("\n\n")
```
